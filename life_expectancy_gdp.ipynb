```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv("all_data.csv")
df.head()
```

```python
# Check the shape of the DataFrame
print(df.shape)

# Check the unique values in 'Country' and 'Year' columns
print(df.Country.unique())
print(df.Year.unique())
```

```python
# Rename the 'Life expectancy at birth (years)' column to 'LEABY'
df = df.rename(columns={"Life expectancy at birth (years)": "LEABY"})
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(df.GDP, kde=False)
plt.xlabel("GDP in Trillions of U.S. Dollars")
plt.title("Distribution of GDP")
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.distplot(df.LEABY, kde=False)
plt.xlabel("Life expectancy at birth (years)")
plt.title("Distribution of Life Expectancy")
plt.show()
```

```python
df_means = df.groupby('Country').mean().reset_index()
df_means
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="LEABY", y="Country", data=df_means)
plt.xlabel("Life expectancy at birth (years)")
plt.title("Average Life Expectancy by Country")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="GDP", y="Country", data=df_means)
plt.xlabel("GDP in Trillions of U.S. Dollars")
plt.title("Average GDP by Country")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
sns.violinplot(ax=axes[0], x='GDP', y='Country', data=df)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
sns.violinplot(ax=axes[1], x='LEABY', y='Country', data=df)
axes[1].set_xlabel("Life expectancy at birth (years)")
plt.suptitle("Violin Plots of GDP and Life Expectancy by Country")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
sns.swarmplot(ax=axes[0], x='GDP', y='Country', data=df)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
sns.swarmplot(ax=axes[1], x='LEABY', y='Country', data=df)
axes[1].set_xlabel("Life expectancy at birth (years)")
plt.suptitle("Swarm Plots of GDP and Life Expectancy by Country")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=df, x="Year", y="GDP", hue="Country")
plt.xlabel("Year")
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.title("GDP Over Time by Country")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=df, x="Year", y="LEABY", hue="Country")
plt.xlabel("Year")
plt.ylabel("Life Expectancy at Birth (years)")
plt.title("Life Expectancy Over Time by Country")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
g = sns.FacetGrid(df, col="Country", col_wrap=3, sharey=False)
g.map(sns.lineplot, "Year", "GDP")
g.add_legend()
g.set_axis_labels("Year", "GDP in Trillions of U.S. Dollars")
plt.show()
```

```python
g = sns.FacetGrid(df, col="Country", col_wrap=3, sharey=False)
g.map(sns.lineplot, "Year", "LEABY")
g.add_legend()
g.set_axis_labels("Year", "Life Expectancy at Birth (years)")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="LEABY", y="GDP", hue="Country")
plt.xlabel("Life Expectancy at Birth (years)")
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.title("GDP vs Life Expectancy by Country")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
g = sns.FacetGrid(df, col="Country", col_wrap=3, sharex=False, sharey=False)
g.map(sns.scatterplot, "LEABY", "GDP")
g.add_legend()
g.set_axis_labels("Life Expectancy at Birth (years)", "GDP in Trillions of U.S. Dollars")
plt.show()
```
